Problem - 4: Matrix Operations:

• For the following arrays:

A = np.array([[3,4],[7,8]]) and B = np.array([[5,3],[2,1]]);

Prove following with Numpy:
1. Prove A.A−1 = I.
2. Prove AB ̸= BA.
3. Prove (AB)

T = BTAT
.

• Solve the following system of Linear equation using Inverse Methods.

2x − 3y + z = −1
x − y + 2z = −3
3x + y − z = 9

{Hint: First use Numpy array to represent the equation in Matrix form. Then Solve for: AX = B}
• Now: solve the above equation using np.linalg.inv function.{Explore more about ”linalg” function
of Numpy}

```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])
print("Matrix A:")
print(A)
print("\nMatrix B:")
print(B)
print()
#1)
print("1. Prove A.A^(-1) = I:")
try:
    A_inv = np.linalg.inv(A)
    identity_check = np.dot(A, A_inv)
    print("A inverse:")
    print(A_inv)
    print("A × A^(-1):")
    print(identity_check)
    print("This is approximately identity matrix I (with small rounding errors)")
except np.linalg.LinAlgError:
    print("Matrix A is singular, cannot compute inverse")
print()
#2)
print("2. Prove AB ≠ BA:")
AB = np.dot(A, B)
BA = np.dot(B, A)
print("A × B:")
print(AB)
print("B × A:")
print(BA)
print("Are they equal?", np.array_equal(AB, BA))
print()
#3)
print("3. Prove (AB)^T = B^T A^T:")
AB_transpose = AB.T
B_transpose_A_transpose = np.dot(B.T, A.T)
print("(AB)^T:")
print(AB_transpose)
print("B^T A^T:")
print(B_transpose_A_transpose)
print("Are they equal?", np.array_equal(AB_transpose, B_transpose_A_transpose))
print()


# Solving linear equations
print("Solving system of linear equations:")
print("2x - 3y + z = -1")
print("x - y + 2z = -3")
print("3x + y - z = 9")
A_coeff = np.array([[2, -3, 1],
                    [1, -1, 2],
                    [3, 1, -1]])
B_const = np.array([[-1], [-3], [9]])

print("\nCoefficient matrix A:")
print(A_coeff)
print("\nConstant matrix B:")
print(B_const)
try:
    A_inv_coeff = np.linalg.inv(A_coeff)
    X = np.dot(A_inv_coeff, B_const)
    print("\nSolution using inverse method:")
    print(f"x = {X[0][0]:.2f}")
    print(f"y = {X[1][0]:.2f}")
    print(f"z = {X[2][0]:.2f}")

    verification = np.dot(A_coeff, X)
    print("A × X should equal B:")
    print("A × X =", verification.flatten())
    print("B =", B_const.flatten())
except np.linalg.LinAlgError:
    print("Matrix is singular, cannot solve using inverse method")
```
